```python
#Boxplot
plt.figure(figsize=(12,6))
sns.boxplot(data=numeric_df)
plt.title("Boxplot with Numeric Summary")

for i, column in enumerate(numeric_df.columns):    #give number to each item like age is 0

    column_data = numeric_df[column]

    # Calculate statistics
    Q1 = column_data.quantile(0.25)
    Q2 = column_data.quantile(0.50)  # Median
    Q3 = column_data.quantile(0.75)
    IQR = Q3 - Q1

    lower_whisker = column_data[column_data >= Q1 - 1.5 * IQR].min()
    upper_whisker = column_data[column_data <= Q3 + 1.5 * IQR].max()

    outliers = column_data[(column_data < Q1 - 1.5*IQR) | (column_data > Q3 + 1.5*IQR)]       # Outliers


    x_pos = i  #Write the text on the boxplot that belongs to this column.

    # Label whiskers
    plt.text(x_pos, lower_whisker, f"{lower_whisker:.2f}",
             ha='center', va='top', color='blue', fontsize=9)

    plt.text(x_pos, upper_whisker, f"{upper_whisker:.2f}",
             ha='center', va='bottom', color='blue', fontsize=9)

    # Label median
    plt.text(x_pos, Q2, f"{Q2:.2f}",
             ha='center', va='center', color='red', fontsize=9)

    # Label outliers
    for out in outliers:
        plt.text(x_pos, out, f"{out:.2f}",        #2f: show the number with only 2 digit after the decimal
                 ha='center', va='center', color='red', fontsize=9)
plt.tight_layout()
plt.show()
```

```python
#Barplot
plt.figure(figsize=(10,5))
variance.plot(kind='bar', color='skyblue')
plt.title("Variance of Numeric Columns")
plt.ylabel("Variance")
plt.show()
```

In dataset, the variance and standard deviation show how spread out each column is, and these numerical values match what you see in the histogram and boxplot. For Age, the standard deviation is around 16.46 and the variance is about 270.92, which are both relatively small. This means Age values are not extremely spread out, so the histogram looks narrow and the boxplot has a small box and short whiskers. In contrast, Annual_Premium has a much larger spread, with a standard deviation of about 1307 and a variance of over 1.7 million. This large spread appears in the histogram as a wide distribution and in the boxplot as a stretched box with longer whiskers. Claim_Amount has the largest variation, with a very high standard deviation (≈ 5882) and extremely large variance (≈ 34 million), which indicates the values are widely scattered with some very high claims. This results in a histogram that is very wide and a boxplot with a tall box, long whiskers, and possibly many outliers. In summary, low variance and standard deviation create narrow, compact plots, while high variance and standard deviation create wider, more spread-out histograms and boxplots.   in short, Low variance/std dev (Age) creates tight, compact plots; high variance/std dev (Premium & Claim Amount) creates wide, stretched plots with longer whiskers and more outliers.

lets compute systematic calculate:

```python
calc = pd.DataFrame({
    "Mean" : df[numeric_columns].mean(),
    "variance" : df[numeric_columns].var().round(2),
    "standard_deviation" : df[numeric_columns].std().round(2),
    "range" : df[numeric_columns].max() - df[numeric_columns].min(),
    "Cofficient of Variation" : df[numeric_columns].std() / df[numeric_columns].mean()

})
calc
```

This code makes a table that gives important numbers for each numeric column in your data. It calculates:
*   Mean → the average value
*   Variance → how spread out the numbers are
*   Standard deviation → how far numbers usually are from the average
*   Range → the difference between the biggest and smallest value
*   Coefficient of variation → how much the data changes compared to the average

All these values together help you quickly understand how the numbers in your dataset behave — whether they are close together, very spread out, or highly variable. The final table (calc) shows all these results in one place.
