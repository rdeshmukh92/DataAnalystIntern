Step 7 — Visualization Tasks(Read Instrutions PDF)

Create the following 4 charts (Google Colab -Python):
1. Bar Chart — Gender vs Units Sold
2. Line Chart — Daily Total Sales
3. Histogram — Customer Age
4. Pie Chart — Customer Type Distribution

```python
#Bar Chart — Gender vs Units Sold
plt.figure(figsize=(8, 8))
ax = sns.barplot(x = df["Gender"], y = df["Units_Sold"], estimator= sum, hue= df["Gender"], errorbar= None)
for container in ax.containers:
  ax.bar_label(container, fontsize= 8);

plt.show()
```

```python
from IPython.core.magics import auto
#Line Chart — Daily Total Sales

daily_sales = df.groupby('Purchase_Date')['Total_Sales'].sum().reset_index()


plt.figure(figsize=(8, 4))
sns.lineplot(x = daily_sales["Purchase_Date"], y = daily_sales["Total_Sales"], legend= auto, marker= 'o')
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,6))
ax = sns.histplot(df["Customer_Age"], bins= [0,20, 40, 60, 80], edgecolor = "black")
for container in ax.containers:
  ax.bar_label(container, fontsize= 8);
plt.show()
```

```python
#Pie Chart — Customer Type Distribution
plt.figure(figsize=(8,6))
df["Customer_Type"].value_counts().plot(kind= "pie", autopct='%1.1f%%')
 #referred:  https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html#sphx-glr-gallery-pie-and-polar-charts-pie-features-py
plt.ylabel("")
plt.show()
```
