#Objective of the Task

The objective of this task is to develop a machine learning model that can predict the likelihood of heatwave conditions in major Indian states for the year 2026 using historical climate data from the past 20 years (2006–2025).

The task aims to:

Analyze historical temperature and climate patterns

Identify key factors contributing to heatwaves

Build and optimize a Random Forest classification model

Perform full machine learning workflow in Google Colab

Generate state-wise heatwave risk predictions for 2026

Provide insights that can support early warning and planning

#Why This Task is Important

Heatwaves are becoming more frequent and severe in India due to:

Climate change and rising global temperatures

Urban heat island effects

Reduced rainfall and changing weather patterns

Predicting heatwaves in advance helps:

Reduce heat-related illnesses and deaths

Support India Meteorological Department (IMD) early warnings

Assist government agencies in disaster preparedness

Plan electricity, water, and healthcare resources

Help farmers and urban planners take preventive measures

This project demonstrates how machine learning can support real-world climate risk management and public policy.


#Dataset Overview


Time Period: 2006–2025 (Daily)

States: 12 major heatwave-prone Indian states

Total Size: ~87,600+ rows (20 years × 365 days × 12 states)

| Column             | Meaning                                |
| ------------------ | -------------------------------------- |
| Date               | Daily date                             |
| Year               | Year (2006–2025)                       |
| State              | State name                             |
| Avg_Temperature    | Average daily temperature (°C)         |
| Max_Temperature    | Maximum daily temperature (°C)         |
| Humidity           | Relative humidity (%)                  |
| Rainfall           | Daily rainfall (mm)                    |
| Wind_Speed         | Wind speed (km/h)                      |
| Urbanization_Index | Urban heat effect (0–1)                |
| Heatwave           | Target (1 = Heatwave, 0 = No Heatwave) |
